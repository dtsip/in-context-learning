# Core plot

```python
def valid_standard(r):
    return (
        re.match("Transformer$", r.model) # | LSTM
        and r.n_dims == 20
        and r.task == "linear_regression"
        and "flat" not in r.run_name
        and "tasks" not in r.run_name
        and "trial" not in r.run_name
        and "training" not in r.run_name
    )

def rename_main_lstm(model_name, r):
    if "LSTM 5120" in model_name:
        return "LSTM"
    else:
        return model_name

core_metrics = collect_results(run_dir, df, valid_standard, rename_model=rename_main_lstm)
```

# Capacity

```python
def valid_capacity(r):
    return (
        re.match("Transformer", r.model)
        and r.task == "linear_regression"
        and "flat" not in r.run_name
        and "tasks" not in r.run_name
        and "training" not in r.run_name
    )

def dims_rename_eval(eval_name, r):
    return eval_name + f"_dims={r.n_dims}"

def trial_rename_model(model_name, r):
    if "trial" in r.run_name:
        return f"{model_name}, trial {r.run_name[-1:]}"
    else:
        return f"{model_name}, trial 1" 

capacity_metrics = collect_results(
    run_dir, df, valid_capacity, rename_eval=dims_rename_eval, rename_model=trial_rename_model
)
```

```python
evals = [
    "standard", "random_quadrants", "skewed_normalize=True", "half_subspace_normalize=True",
    "noisyLR_std=scaling_renorm=True", "orthogonal_train_test", "overlapping_train_test",
    "scale-x=2", "scale-y=2",
]
models = ["Transformer-xs", "Transformer-small", "Transformer"]
param_count = [3379456, 7648640, 22395136]
k = 2
plot = plt.loglog

for i, eval_name in enumerate(evals):
    plt.figure(figsize=(3,3))
    trivial = 1.0 if "noisy" not in eval_name else (1+1/20)
    plt.axhline(trivial, ls='--', color="gray")

    for dims in [10, 20, 30, 40, 50]:
        if "orthogonal" in eval_name:
            idx = dims - 1
        else:
            idx = dims * k
        lines = [[], [], []]
        for model in models:
            res = []
            for trial in ["1", "2", "3"]:
                k1 = f"{eval_name}_dims={dims}"
                k2 = f"{model}, trial {trial}"
                if k1 not in capacity_metrics or k2 not in capacity_metrics[k1]:
                    print(k1, k2)
                res.append(capacity_metrics[k1][k2]["mean"][idx])
            for l, n in zip(lines, sorted(res)):
                l.append(n)
        plot(param_count, lines[1], marker="o", label=dims, lw=2)
        plt.fill_between(param_count, lines[0], lines[2], alpha=.3)

    #plt.xlim(-.1, 2.1)
    plt.xlabel("number of parameters")
    plt.tick_params(axis='x', which='minor', bottom=False, labelbottom=False)
    plt.xticks(param_count, ["3.4M", "7.6M", "22.4M"])

    plt.ylim(1e-4, 15)
    if i in [0, 3, 6]:
        plt.ylabel("squared error")
        #plt.ylim(5e-6, 5)
    if i in [2]:
        legend = plt.legend(title="dimensions", loc="upper left", bbox_to_anchor=(1, 1))
        for line in legend.get_lines():
            line.set_linewidth(3)
    if SAVE:
        plt.savefig(f"figures/capacity_{eval_name.split('_')[0]}.pdf", bbox_inches='tight')
    plt.show()
```

```python
plt.figure(figsize=(0,0))
for i, d in enumerate([10, 20, 30, 40, 50]):
    plt.plot([1], '-', label=d, color=palette[i], lw=3)
plt.legend(ncol=7, fontsize=12, title="dimensions")
plt.axis('off')
plt.savefig(f"figures/capacity_legend.pdf", bbox_inches='tight')
plt.show()
```

# Finite

```python
def valid_finite(r):
    return (
        "Transformer" == r.model
        and r.n_dims == 20
        and r.task == "linear_regression"
        and "flat" not in r.run_name
        and "trial" not in r.run_name
        and "training" not in r.run_name
    )

def finite_rename_model(model_name, r):
    tasks = r.num_tasks
    if tasks < 0:
        tasks = 32000000
    model_name += f"_{tasks}"
    return model_name

finite_metrics = collect_results(run_dir, df, valid_finite, rename_model=finite_rename_model)
```

```python
TASKS = [100, 1000, 10000, 100000, 32000000]
plt.figure()
color = 0
for idx in [10, 20, 30, 40]:
    low, mid, high = [], [], []
    for tasks in TASKS:
        vs = finite_metrics["standard"][f"Transformer_{tasks}"]
        low.append(vs["bootstrap_low"][idx])
        high.append(vs["bootstrap_high"][idx])
        mid.append(vs["mean"][idx])
    
    plt.axhline(1.0, ls="--", color="grey")
    plt.semilogx(TASKS, mid, "-", color=palette[color], label=f"{idx}", lw=2, marker='o')
    color += 1
    plt.fill_between(TASKS, low, high, alpha=.3)
    legend = plt.legend(title="#in-context examples", loc="upper left", bbox_to_anchor=(1, 1))
    for line in legend.get_lines():
        line.set_linewidth(3)
#plt.ylabel("squared error")
plt.ylabel("")
plt.xlabel("#weight vectors used for training")
plt.xticks(TASKS, ["100", "1K", "10K", "100K", "32M"])
plt.gcf().set_size_inches(4,3)
if SAVE:
    plt.savefig(f"figures/finite.pdf", bbox_inches='tight')
plt.show()

#plt.tight_layout()
```

# Scale

```python
for model_name, model in [
    ("transformer", "Transformer"),
    #("lstm_wide", "LSTM"),
    #("lstm_deep", "LSTM 1024 x 4"),
]:
    for i, axis in enumerate(["x", "y"]):
        fig, ax = plt.subplots(1, 1, figsize=(4, 3))
        color = 0
        ax.axhline(1, ls='--', color="grey")
        for j, scale in enumerate([0.333, 0.5, 1, 2, 3]):
            if scale == 1:
                eval_name = "standard"
            else:
                eval_name = f"scale-{axis}={scale}"
            vs = core_metrics[eval_name][model]
            vs_norm = {}
            for k in vs:
                vs_norm[k] = [v / scale**2 for v in vs[k]]
            label = {0.5: "1/2",0.333: "1/3",1: "1",2: "2",3: "3"}[scale]
            ax.plot(vs_norm["mean"], "-", label=label, color=palette[color], lw=2)
            color += 1
            low = vs_norm["bootstrap_low"]
            high = vs_norm["bootstrap_high"]
            ax.fill_between(range(len(low)), low, high, alpha=.3)
        vs = core_metrics["standard"]["Least Squares"]
        vs_norm = {}
        for k in vs:
            vs_norm[k] = [v for v in vs[k]]
        ax.plot(vs_norm["mean"], "--", color=palette[color], label="Least Squares")
        color += 1
        low = vs_norm["bootstrap_low"]
        high = vs_norm["bootstrap_high"]
        ax.fill_between(range(len(low)), low, high, alpha=.3)

        ax.set_ylim(-.1, 1.1)
        ax.set_xlabel("in-context examples")
        if i == 0:
            ax.set_ylabel("squared error")
        if i == 1:
            legend = ax.legend(title="scale")
            for line in legend.get_lines():
                line.set_linewidth(3)
        plt.savefig(f"figures/scale{axis}_{model_name}.pdf", bbox_inches='tight')
        #plt.tight_layout()
    plt.show()
```

# trials

```python
trial_metrics = {}
for n_dims in [10, 20, 30, 40, 50]:
    def valid_trial(r):
        return (
            re.match("Transformer$", r.model)
            and r.n_dims == n_dims
            and r.task == "linear_regression"
            and "flat" not in r.run_name
            and "tasks" not in r.run_name
            and "training" not in r.run_name
        )

    trial_metrics[n_dims] = collect_results(run_dir, df, valid_trial, rename_model=trial_rename_model)
```

```python
models = [
    "Transformer, trial 1", "Transformer, trial 2", "Transformer, trial 3", "Least Squares"
]
metrics = [
    "standard", "random_quadrants", "skewed_normalize=True",
    "half_subspace_normalize=True",  "noisyLR_std=scaling_renorm=True", "orthogonal_train_test",
    "overlapping_train_test", "scale-x=2", "scale-y=2",
]

for n_dims in [10, 20, 30, 40, 50]:
    for metric in metrics:
        trivial = 1.0 if "noisy" not in metric else (1+1/20)
        fig, ax = basic_plot(trial_metrics[n_dims][metric], models=models, trivial=trivial)
        ax.get_legend().remove()
        if n_dims != 10:
            plt.ylabel("")
        if "orthogonal" in metric:
            plt.xlim(-0.1 * n_dims, 0.9*n_dims)
        else:
            plt.xlim(-0.1 * n_dims, 2.1*n_dims)
            plt.xticks([0, n_dims, 2*n_dims])
        plt.title(f"{n_dims} dimensions")
        plt.gcf().set_size_inches(2, 2)
        plt.savefig(f"figures/trial_{metric.split('_')[0]}_{n_dims}.pdf", bbox_inches='tight')
#plt.tight_layout()
```

```python
plt.figure(figsize=(0,0))
for i, m in enumerate(models):
    plt.plot([1], '-', label=m, color=palette[i], lw=3)
plt.legend(ncol=4, fontsize=12)
plt.axis('off')
plt.savefig(f"figures/trial_legend.pdf", bbox_inches='tight')
plt.show()
```

# Curriculum

```python
def curr_rename_model(model_name, r):
    if "flat" in r.run_name:
        return f"{model_name}, no curriculum"
    else:
        return f"{model_name}" 

curr_metrics = {}
for n_dims in [10, 20, 30, 40, 50]:
    def valid_curr(r):
        return (
            re.match("Transformer$", r.model)
            and r.n_dims == n_dims
            and r.task == "linear_regression"
            and "trial" not in r.run_name
            and "tasks" not in r.run_name
            and "training" not in r.run_name
        )

    curr_metrics[n_dims] = collect_results(run_dir, df, valid_curr, rename_model=curr_rename_model)
```

```python
metrics = [
    "standard", "random_quadrants", "skewed_normalize=True",
    #"half_subspace_normalize=True",  "noisyLR_std=scaling_renorm=True", "orthogonal_train_test",
    #"overlapping_train_test", "scale-x=2", "scale-y=2",
]

models = [
    "Transformer", "Transformer, no curriculum", "Least Squares"
]

for n_dims in [20]:
    for metric in metrics:
        fig, ax = basic_plot(curr_metrics[n_dims][metric], models=models)
        ax.get_legend().remove()
        if metric != "standard":
            plt.ylabel("")
        plt.xlim(-0.1 * n_dims, 2.1*n_dims)
        #plt.xticks([0, n_dims, 2*n_dims])
        #plt.title(f"{n_dims} dimensions")
        plt.gcf().set_size_inches(3, 3)
        if "skewed" in metric:
            plt.legend(loc="upper left", bbox_to_anchor=(.1, 1))
        plt.savefig(f"figures/curr_{metric.split('_')[0]}_{n_dims}.pdf", bbox_inches='tight')
```
